```python
print("Body Mass Index (BMI) Calculation Program")

height = float(input("Enter height (m): "))
# Using float for weight is better (e.g., someone might be 75.5 kg)
weight = float(input("Enter weight (kg): "))

# Calculate BMI
bmi = weight / (height ** 2)

# Print the number nicely formatted to 2 decimal places
print(f"\nYour BMI: {bmi:.2f}")

# Standard BMI Categories
if bmi < 18.5:
    print("Category: Underweight")
elif 18.5 <= bmi < 25:
    print("Category: Normal Weight")
elif 25 <= bmi < 30:
    print("Category: Overweight")
else:
    # Any number 30 or higher falls here
    print("Category: Obese")
```

```python
print("--- Python Number Tool ---")

while True:
    print("\nSelect an operation:")
    print("1. List numbers 1-100")
    print("2. List Even numbers 1-100")
    print("3. List Odd numbers 1-100")
    print("4. List numbers divisible by 3 AND 5 (from 1-100)")
    print("5. Count up to your own number")
    print("6. Exit")

    choice = input("Enter choice (1-6): ")

    print("-" * 20) # Prints a divider line

    # Option 1: All numbers 1-100
    if choice == '1':
        for i in range(1, 101):
            print(i)

    # Option 2: Even numbers (Start at 2, step by 2)
    elif choice == '2':
        for i in range(2, 101, 2):
            print(i)

    # Option 3: Odd numbers (Start at 1, step by 2)
    elif choice == '3':
        for i in range(1, 101, 2):
            print(i)

    # Option 4: Divisible by 3 AND 5 (Multiples of 15)
    elif choice == '4':
        for i in range(1, 101):
            # Using 'and' means it must satisfy BOTH conditions
            if i % 3 == 0 and i % 5 == 0:
                print(i)

    # Option 5: User entered number
    elif choice == '5':
        user_num = input("Enter your limit: ")
        # We convert to int inside the range
        for i in range(1, int(user_num) + 1):
            print(i)

    # Option 6: Stop the loop
    elif choice == '6':
        print("Goodbye!")
        break # Breaks the 'while True' loop

    # Invalid input handling
    else:
        print("Invalid choice. Please type 1, 2, 3, 4, 5, or 6.")
```

```python
import random
import math

# Taking Inputs
lower = int(input("Enter Lower bound: "))
upper = int(input("Enter Upper bound: "))

# Generating random number between the lower and upper
x = random.randint(lower, upper)

# Calculate the optimal number of guesses
# We convert it to int() so we don't get decimals in the print statement
total_chances = round(math.log(upper - lower + 1, 2))

print("\nYou have only", total_chances, "chances to guess the integer!\n")

# Initializing the number of guesses
count = 0

# Loop continues while count is less than the allowed chances
while count < total_chances:
    count += 1

    # Taking guessing number as input
    guess = int(input("Guess a number: "))

    # Condition testing
    if x == guess:
        print("Congratulations you did it in", count, "try")
        # Once guessed, loop will break
        break
    elif x > guess:
        print("You guessed too small!")
    elif x < guess:
        print("You guessed too high!")

# If Guessing is more than required guesses
if count >= total_chances and x != guess:

    print("\nThe number was", x)
    print("Better Luck Next time!")
```

```python
import time

def countdown(time_sec):
    while time_sec:
        mins, secs = divmod(time_sec, 60)
        timeformat = '{:02d}:{:02d}'.format(mins, secs)
        print(timeformat, end='\r')
        time.sleep(1)
        time_sec -= 1

    print("stop")

countdown(5)
```

```python
def get_permutation(string, i=0):

    if i == len(string):   	 
        print("".join(string))

    for j in range(i, len(string)):

        words = [c for c in string]
   
        # swap
        words[i], words[j] = words[j], words[i]
   	 
        get_permutation(words, i + 1)

print(get_permutation('why'))
```

```python
shopping_list = []

print("--- My Shopping List ---")

while True:
    print("\nOptions: (1) Add  (2) Remove  (3) View  (4) Exit")
    choice = input("Enter choice: ")

    if choice == '1':
        item = input("What do you want to add? ")
        shopping_list.append(item)
        print("Added:", item)

    elif choice == '2':
        item = input("What do you want to remove? ")
        if item in shopping_list:
            shopping_list.remove(item)
            print("Removed:", item)
        else:
            print("Error: That item is not in your list.")

    # --- THIS IS THE NEW PART ---
    elif choice == '3':
        print("\n--- Your Current List ---")
        # i is the counter (starts at 0), item is the text
        for i, item in enumerate(shopping_list):
            # We print (i + 1) because computers start at 0
            print(i + 1, ".", item) 
    # ----------------------------

    elif choice == '4':
        print("Goodbye!")
        break
```

```python
def get_fahrenheit(celsius):
    result = (celsius * 1.8) + 32
    # Hand the result back to whoever asked for it
    return result

# --- Main Program ---

# 1. We call the function and STORE the answer in a variable
temp_ny = get_fahrenheit(10)
temp_la = get_fahrenheit(30)

# 2. Now we can do things with those variables
print("New York is:", temp_ny)
print("Los Angeles is:", temp_la)

# 3. We can even do logic on the returned value!
if temp_la > 80:
    print("It is a hot day in LA.")
```

```python
# 1. Define the function
def convert_to_fahrenheit(celsius):
    # The math formula
    fahrenheit = (celsius * 1.8) + 32
    
    # Using the Comma Method
    print(celsius, "C is equal to", fahrenheit, "F")

# 2. Now we test it (Call the function)
print("--- Temperature Test ---")

# Test 1: Water freezing
convert_to_fahrenheit(0)

# Test 2: A hot day
convert_to_fahrenheit(30)

# Test 3: Water boiling
convert_to_fahrenheit(100)
```

```python
print("--- Safe Division Calculator ---")

try:
    # 1. We put the risky code inside the 'try' block
    num1 = int(input("Enter top number: "))
    num2 = int(input("Enter bottom number: "))
    
    result = num1 / num2
    
    print("Result:", result)

except ValueError:
    # 2. This runs if they type text like "apple"
    print("Error: You must enter a valid number (0-9).")

except ZeroDivisionError:
    # 3. This runs if they try to divide by 0
    print("Error: You cannot divide by zero!")

except:
    # 4. This catches ANY other unknown error
    print("Error: Something went wrong.")
```
